```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
labels = ['fear','greed']
label_embeddings = model.encode(labels)
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/small_merged_labeled_cleaned.csv", sep='|', index_col=False)
df = df.astype({"clean_body": str})
```

```python
from sentence_transformers import util
result_dict = {}
for index, row in df.iterrows():
    cosine_scores = util.cos_sim(model.encode(row["clean_body"]), label_embeddings)
    result_dict[index] = {labels[i]: cosine_scores[0][i].item() for i in range(len(labels))}
```

```python
matched = 0
for i in range(len(result_dict)):
    if df.iloc[i].label == 'fear' if result_dict[i]['fear']>result_dict[i]['greed'] else 'greed':
        matched+=1
```
